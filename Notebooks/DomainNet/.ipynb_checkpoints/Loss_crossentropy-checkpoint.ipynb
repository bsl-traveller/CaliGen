```python
import pandas as pd
import os
from utils import *
```

```python
res_columns = ['Calibrated on', 'clipart', 'infograph', 'painting', 'quickdraw', 'real', 'sketch', 'Average']
```

```python
best_rho_er_in = best_rho_calib('Error', 'in', 'crossentropy')
best_rho_ece_in = best_rho_calib('ECE', 'in', 'crossentropy')
```

```python
error_file_in = os.path.join('..', '..', 'results', 'DomainNet', 'crossentropy', 'in_Error_mean.csv')
error_in = pd.read_csv(error_file_in)
error_in = get_res_df(error_in, best_rho_er_in, best_rho_ece_in)
```

```python
res_mean_in = error_in[MAIN_COLUMNS+['Domain']][(error_in['Valid'] == False) & (error_in['Train'] == False)].groupby(['Domain']).mean().T
res_std_in = error_in[MAIN_COLUMNS+['Domain']][(error_in['Valid'] == False) & (error_in['Train'] == False)].groupby(['Domain']).std().T
```

```python
res_mean_in['Average'] = res_mean_in[['clipart', 'infograph', 'painting', 'quickdraw', 'real', 'sketch']].mean(axis=1)
```

```python
res_std_in['std'] = get_std(res_mean_in, res_std_in)
```

```python
res_mean_in = res_mean_in.round(2)
res_std_in = res_std_in.round(2)
```

```python
for domain in ['clipart', 'infograph', 'painting', 'quickdraw', 'real', 'sketch']:
    res_mean_in[domain] = res_mean_in[domain].astype(str) + '+' + res_std_in[domain].astype(str)
res_mean_in['Average'] = res_mean_in['Average'].astype(str) + '+' + res_std_in['std'].astype(str)
```

```python
res_mean_in['Calibrated on'] = ['in'] * len(res_mean_in)
res_mean_in.rename_axis('Method', axis='columns', inplace=True)
```

```python
res_mean_in[res_columns]
```

```python
best_rho_er_out = best_rho_calib('Error', 'out', 'crossentropy')
best_rho_ece_out = best_rho_calib('ECE', 'out', 'crossentropy')
```

```python
error_file_out = os.path.join('..', '..', 'results', 'DomainNet', 'crossentropy', 'out_Error_mean.csv')
error_out = pd.read_csv(error_file_out)
error_out = get_res_df(error_out, best_rho_er_out, best_rho_ece_out)
```

```python
res_mean_out = error_out[MAIN_COLUMNS+['Domain']][(error_out['Valid'] == False) & (error_out['Train'] == False)].groupby(['Domain']).mean().T
res_std_out = error_out[MAIN_COLUMNS+['Domain']][(error_out['Valid'] == False) & (error_out['Train'] == False)].groupby(['Domain']).std().T
```

```python
res_mean_out['Average'] = res_mean_out[['clipart', 'infograph', 'painting', 'quickdraw', 'real', 'sketch']].mean(axis=1)
res_std_out['std'] = get_std(res_mean_out, res_std_out)
```

```python
res_mean_out = res_mean_out.round(2)
res_std_out = res_std_out.round(2)
```

```python
for domain in ['clipart', 'infograph', 'painting', 'quickdraw', 'real', 'sketch']:
    res_mean_out[domain] = res_mean_out[domain].astype(str) + '+' + res_std_out[domain].astype(str)
res_mean_out['Average'] = res_mean_out['Average'].astype(str) + '+' + res_std_out['std'].astype(str)
```

```python
res_mean_out['Calibrated on'] = ['out'] * len(res_mean_out)
res_mean_out.rename_axis('Method', axis='columns', inplace=True)
```

```python
res_mean_out[res_columns]
```

```python
res = pd.concat([res_mean_in[res_columns], res_mean_out[res_columns]])
res
```
