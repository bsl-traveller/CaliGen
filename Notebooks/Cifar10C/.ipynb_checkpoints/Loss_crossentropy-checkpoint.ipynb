```python
import pandas as pd
import os
from utils import *
```

```python
res_columns = ['Train Filter', 'Calib Filter', 'Test Filter (1)', 'All Filter (2)', 'All Filter (3)', 'All Filter (4)', 'All Filter (5)']
```

```python
best_rho_er_in = best_rho_calib('Error', 'in', 'crossentropy', '4filter_1')
best_rho_ece_in = best_rho_calib('ECE', 'in', 'crossentropy', '4filter_1')
```

```python
best_rho_er_in
```

```python
ece_file_in = os.path.join('..', '..', 'results', 'cifar10c', 'crossentropy', '4filter_1', 'in_ECE_mean.csv')
ece_in = pd.read_csv(ece_file_in)
ece_in = get_rho_df(ece_in, best_rho_er_in, best_rho_ece_in)
```

```python
filters_calib = ['fog', 'contrast', 'elastic_transform', 'saturate']

filters_train = ['gaussian_noise', 'brightness', 'pixelate', 'gaussian_blur']

filter_rest = ['shot_noise', 'impulse_noise', 'defocus_blur', 'glass_blur', 'zoom_blur', 'jpeg_compression', 'speckle_noise']
```

```python
train_f, valid_f, rest_f = get_level_filters('4filter_1', filters_train, filters_train, filter_rest)
```

```python
result_in, result_std_in = get_res_df(res_columns, ece_in, '4filter_1', train_f, valid_f, rest_f)
```

```python
result_in.rename_axis('Method', axis='columns', inplace=True)
result_std_in['std'] = get_std_c(result_in, result_std_in, res_columns)
result_std_in = result_std_in.round(2)
```

```python
for column in res_columns:
    result_in[column] = result_in[column].astype(str) + '+' + result_std_in[column].astype(str)
result_in['Average'] = result_in['Average'].astype(str) + '+' + result_std_in['std'].astype(str)
```

```python
result_in['Calibrated on'] = ['in'] * len(result_in)
```

```python
best_rho_er_out = best_rho_calib('Error', 'out', 'crossentropy', '4filter_1')
best_rho_ece_out = best_rho_calib('ECE', 'out', 'crossentropy', '4filter_1')
```

```python
ece_file_out = os.path.join('..', '..', 'results', 'cifar10c', 'crossentropy', '4filter_1', 'out_ECE_mean.csv')
ece_out = pd.read_csv(ece_file_out)
ece_out = get_rho_df(ece_out, best_rho_er_out, best_rho_ece_out)
```

```python
result_out, result_std_out = get_res_df(res_columns, ece_out, '4filter_1', train_f, valid_f, rest_f)
```

```python
result_out.rename_axis('Method', axis='columns', inplace=True)
result_std_out['std'] = get_std_c(result_out, result_std_out, res_columns)
result_std_out = result_std_out.round(2)
```

```python
for column in res_columns:
    result_out[column] = result_out[column].astype(str) + '+' + result_std_out[column].astype(str)
result_out['Average'] = result_out['Average'].astype(str) + '+' + result_std_out['std'].astype(str)
```

```python
result_out['Calibrated on'] = ['out'] * len(result_out)
```

```python
res = pd.concat([result_in, result_out])
```

```python
res_col = ['Calibrated on', 'Train Filter', 'Calib Filter', 'Test Filter (1)', 'All Filter (2)', 'All Filter (3)', 'All Filter (4)', 'All Filter (5)', 'Average']
res[res_col]
```
